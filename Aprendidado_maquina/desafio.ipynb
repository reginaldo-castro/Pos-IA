```python
import cv2
import numpy as np

video = cv2.VideoCapture(path + 'mont_russa.mp4')
cont_frames = 0

lim_ini = np.array([0, 100, 100])
lim_fim = np.array([30, 255, 255])

while(True):
    ret, frame = video.read()
    if ret:
        nome = path + '_' + "{0:03}".format(cont_frames) + '.jpg' 
        print(nome)
        cv2.imwrite(nome, frame)
        cont_frames += 1
        im_ar_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        templ_cor = cv2.inRange(im_ar_hsv, lim_ini, lim_fim)
        saida = cv2.bitwise_and(frame, frame, mask=templ_cor)

    else:
        break
print('Fim da geração.')
```

```python
cv2.imshow('Resultado', np.hstack((nome, saida)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

path = "imagens_am/"

video = cv2.VideoCapture(path + 'mont_russa.mp4')
cont_frames = 0

lim_ini = np.array([0, 100, 100])
lim_fim = np.array([30, 255, 255])

cor_vermelha = False

while True:
    ret, frame = video.read()
    
    if ret:
        cont_frames += 1
        
        # Converte o frame para o espaço de cores HSV
        im_ar_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        # Cria uma máscara para a cor vermelha
        templ_cor = cv2.inRange(im_ar_hsv, lim_ini, lim_fim)
        
        # Verifique se a máscara tem algum pixel não preto (vermelho detectado)
        if cv2.countNonZero(templ_cor) > 0:
            cor_vermelha = True
            print(f"Cor vermelha detectada no frame {cont_frames}")
            cv2.imshow('Resultado', frame)
            break

        # Aguarde a tecla 'q' ser pressionada para sair
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    else:
        break

# Fecha a janela quando a tecla 'q' for pressionada
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path = "imagens_am/"

video = cv2.VideoCapture(path + 'mont_russa.mp4')
cont_frames = 0

lim_ini = np.array([0, 100, 100])
lim_fim = np.array([50, 255, 255])
 
cor_vermelha_detectada = False

while True:
    ret, frame = video.read()
    
    if ret:
        cont_frames += 1
        
        im_ar_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        print(im_ar_hsv)
        # Cria uma máscara para a cor vermelha
        templ_cor = cv2.inRange(im_ar_hsv, lim_ini, lim_fim)
        print(templ_cor)
        # Verifique se a máscara tem algum pixel não preto (vermelho detectado)
        if cv2.countNonZero(templ_cor) > 0:
            cor_vermelha_detectada = True
            print(f"Cor vermelha detectada no frame {cont_frames}")
            
            # Exibe o quadro usando matplotlib
            plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            plt.show()
            break

        # Aguarde a tecla 'q' ser pressionada para sair
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    else:
        break

# Fecha a janela quando a tecla 'q' for pressionada
cv2.waitKey(0)
cv2.destroyAllWindows()
           
```
